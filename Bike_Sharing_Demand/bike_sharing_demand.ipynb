```python
train = pd.read_csv('/kaggle/input/bike-sharing-demand/train.csv')
train
```

```python
test = pd.read_csv('/kaggle/input/bike-sharing-demand/test.csv')
test
```

```python
train['datetime'] = pd.to_datetime(train['datetime'])
```

```python
train['Year'] = train['datetime'].dt.year
#train['Month'] = train['datetime'].dt.month
train['Day'] = train['datetime'].dt.day
train['Hour'] = train['datetime'].dt.hour
train['Weekday'] = train['datetime'].dt.weekday
#train['Minute'] = train['datetime'].dt.minute
#train['Second'] = train['datetime'].dt.second
train.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16,8))
sns.boxplot(train['Day'], train['count'])
```

```python
weekday_df = train[train['workingday'] == 1]
weekday_df
```

```python
weekday_df = train.query('workingday==1')
weekday_df
```

```python
weekend_df = train.query('workingday!=1')
weekend_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16,8))
sns.boxplot(weekday_df['Year'],weekday_df['count'])
```

```python
plt.figure(figsize=(16,8))
sns.boxplot(weekend_df['Year'],weekend_df['count'])
```

```python
test['datetime'] = pd.to_datetime(test['datetime'])
test['Year'] = test['datetime'].dt.year
#test['Month'] = test['datetime'].dt.month
#test['Day'] = test['datetime'].dt.day
test['Hour'] = test['datetime'].dt.hour
test['Weekday'] = test['datetime'].dt.weekday
test.head(5)
```

```python
train2 = train.drop(columns=['datetime','count','casual','registered','Day'])
train2
```

```python
test2 = test.drop(columns=['datetime'])
test2
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_jobs=-1)
rf.fit(train2, train['count'])
result = rf.predict(test2)
```

```python
rf.feature_importances_
```

```python
pd.Series(rf.feature_importances_, index=train2.columns).sort_values(ascending=False)
```
