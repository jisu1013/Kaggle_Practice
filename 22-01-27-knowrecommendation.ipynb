```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
cat = all_data_2.columns[all_data_2.dtypes == 'object']
cat
```

```python
for i in cat:
    all_data_2[i] = le.fit_transform(list(all_data_2[i]))
all_data_2
```

```python
train_2 = all_data_2[:len(train)]
test_2 = all_data_2[len(train):]
```

```python
#from catboost import CatBoostClassifier

#cbc = CatBoostClassifier(task_type='GPU')
#cbc.fit(train_2, train['knowcode'])
```

```python
from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_valid = train_test_split(train_2, train['knowcode'], test_size=0.2, random_state=42, stratify=train['knowcode'])
```

```python
#y_train.value_counts()
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(tree_method='gpu_hist', learning_rate=0.1, n_estimators=1000)
xgb.fit(x_train, y_train, eval_set=[(x_valid, y_valid)], early_stopping_rounds=15)
```

```python
#from xgboost import plot_importance
#import matplotlib.pyplot as plt
#%matplotlib inline

#fig, ax = plt.subplots(figsize=(10, 12))
xgb.feature_importances_.min()
```

```python
result = xgb.predict(test_2)
result
```

```python
x_result = xgb.predict(x_valid)
x_result
```

```python
from sklearn.metrics import f1_score

f1_score(y_valid, x_result, average='macro')
```
